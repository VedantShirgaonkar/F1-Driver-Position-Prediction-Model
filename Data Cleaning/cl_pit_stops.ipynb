```python
import pandas as pd
```

```python
cl_pit_stops = pd.read_csv('/Users/rahul/Desktop/archive/cleaned_pit_stops.csv')
cl_pit_stops.head()
```

```python
cl_pit_stops.drop(columns='milliseconds',inplace=True)
cl_pit_stops.head()
```

```python
cl_pit_stops['total_pit_stops'] = cl_pit_stops.groupby(['driverId', 'raceId'])['stop'].transform('count')
cl_pit_stops.head()
```

```python
#cl_pit_stops.rename(columns={'lap': 'pit_stops_lap'}, inplace=True)
cl_pit_stops.drop(columns=['stop','lap'],inplace=True)
cl_pit_stops.head()
```

```python
cl_pit_stops.to_csv('cl_pit_stops.csv',index=False)
```
